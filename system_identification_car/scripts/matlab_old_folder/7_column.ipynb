```python
def get_error_circle(data):
    RADIUS = 150
    data_length = data.shape[0]
    error_array = np.zeros(data_length)
    delta_y = data[:,1] - RADIUS
    r_array = np.sqrt(data[:,0]*data[:,0] + delta_y * delta_y)
    return r_array - RADIUS

def get_error_sine(data):
    AMPLITUDE = 3
    WAVELENGTH = 150
    y_standard = AMPLITUDE * np.sin(data[:,0] * 2 * pi /(WAVELENGTH))
    return data[:,1] - y_standard
```

```python
DATA_LENGTH = 20000
x_index = np.linspace(0, DATA_LENGTH-1, DATA_LENGTH, dtype = int)
```

```python
temp_error = get_error_circle(circle_my)
plt.plot(x_index, temp_error)
print(np.mean(np.square(temp_error)))
```

```python
temp_error = get_error_circle(circle_official)
plt.plot(x_index, temp_error)
print(np.mean(np.square(temp_error)))
```

```python
temp_error = get_error_sine(sine_my)
plt.plot(x_index, temp_error)
print(np.mean(np.square(temp_error)))
```

```python
temp_error = get_error_sine(sine_official)
plt.plot(x_index, temp_error)
print(np.mean(np.square(temp_error)))
```
