```python
import numpy as np
from matplotlib import pyplot as plt
from math import *
```

```python
sine_my = np.load("./running_data_sine_my_controller_new2.npy")
plt.plot(sine_my[:,0], sine_my[:,1])
plt.show()
```

```python
sine_official = np.load("./running_data_sine_official_controller_new2.npy")
plt.plot(sine_official[:,0], sine_official[:,1])
plt.show()
```

```python
def get_error_sine(data):
    AMPLITUDE = 2
    WAVELENGTH = 200 + 50 * np.sin(data[:,0] * 2 * pi / (2500.0));
    y_standard = AMPLITUDE * np.sin(data[:,0] * 2 * pi /(WAVELENGTH))
    return data[:,1] - y_standard
```

```python
DATA_LENGTH = 20000
x_index = np.linspace(0, DATA_LENGTH-1, DATA_LENGTH, dtype = int)
```

```python
temp_error = get_error_sine(sine_my)
plt.plot(x_index, temp_error)
print(np.mean(np.square(temp_error)))
```

```python
temp_error = get_error_sine(sine_official)
plt.plot(x_index, temp_error)
print(np.mean(np.square(temp_error)))
```
